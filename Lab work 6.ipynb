```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
sns.set_style('darkgrid')
fig,ax = plt.subplots(figsize = (15,8))
ax.plot(df.index,df.a,color='r',ls = '-.')
ax.plot(df.index, df.b,color = 'orange', linewidth = 10)
ax.plot(df.index, df.c,color = 'yellow', linewidth = 1, marker = 'o')
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig,ax = plt.subplots(3,figsize = (15,5))

ax[0].plot(df.index,df.a,label = 'Line A')
ax[0].legend(loc ='lower right')
ax[1].plot(df.index,df.b,label = 'Line B')
ax[1].legend(loc = 'center left')
ax[2].plot(df.index,df.c)
```
